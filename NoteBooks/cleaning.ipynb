```python
import pandas as pd
import numpy as np
```

```python
path = "E:\\Data Science Salarys\\Data\\raw\\ds_salaries.csv"
df = pd.read_csv(path, sep=",")
```

```python
df.head()
```

```python
df.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
# Renomear colunas para português

df['experience_level'] = df['experience_level'].replace({
    'EN': 'Junior',
    'MI': 'Pleno',
    'SE': 'Senior',
    'EX': 'Executive'
})

df['employment_type'] = df['employment_type'].replace({
    'PT': 'Meio período',
    'FT': 'Tempo integral',
    'CT': 'Contrato',
    'FL': 'Freelancer'
})


df['company_size'] = df['company_size'].replace({
    'S': 'Pequena',
    'M': 'Média',
    'L': 'Grande'
})

df['remote_ratio'] = df['remote_ratio'].replace({
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
})

df.head()
```

```python
# Converter colunas para tipos adequados

df['salary'] = df['salary'].astype('float')
df['salary_in_usd'] = df['salary_in_usd'].astype('float')
df['remote_ratio'] = df['remote_ratio'].astype('category')
df['experience_level'] = df['experience_level'].astype('category')
df['employment_type'] = df['employment_type'].astype('category')
df['company_size'] = df['company_size'].astype('category')

df.info()
```

```python
# Criar nova coluna "Área de Atuação" com base no cargo

regras = {
    'Cientista de Dados': [
        'scientist', 'science', 'machine learning', 'researcher', '3d'
    ],
    'Analista de Dados': [
        'analyst', 'analytics'
    ],
    'Engenheiro de Dados': [
        'engineer', 'etl'
    ],
    'Arquiteto de Dados': [
        'architect'
    ],
    'Gerente de Dados': [
        'manager', 'head', 'specialist'
    ]
}

def categorizar(cargo):
    cargo = str(cargo).lower()
    
    for categoria, palavras in regras.items():
        if any(palavra in cargo for palavra in palavras):
            return categoria
    
    return 'Outros'

df['Área_de_Atuação'] = df['job_title'].apply(categorizar)
```

```python
df.head()
```

```python
# Renomear colunas para português

df = df.rename(columns={'work_year': 'Ano_de_Trabalho', 
                        'experience_level': 'Nivel_de_Experiencia', 
                        'employment_type': 'Tipo_de_Emprego', 
                        'job_title': 'Cargo', 
                        'salary': 'Salario', 
                        'salary_currency': 'Moeda_do_Salario', 
                        'salary_in_usd': 'Salario_em_USD', 
                        'employee_residence': 'Residencia_do_Funcionario', 
                        'remote_ratio': 'Tipo_de_Contrato', 
                        'company_location': 'Localizacao_da_Empresa', 
                        'company_size': 'Tamanho_da_Empresa'})
```

```python
# Salvar o DataFrame limpo em um novo arquivo CSV

df.to_csv("E:\\Data Science Salarys\\Data\\cleaned\\ds_salaries_limpo.csv", sep=',', decimal=',')
```
